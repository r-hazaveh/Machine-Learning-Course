```python
app.set_index('Date')   #answer 6
```

```python
#print(app.Date.nunique() == app.shape[0])
# The duplicated method in pandas is used to identify duplicate rows in a DataFrame or Series. It returns a boolean Series where True indicates that the row is a duplicate of a previous row and False indicates that it is not.
app.index.duplicated()
```

```python
app.index.duplicated().sum()
```

```python
app_sorted = app.sort_values(by = 'Date', ascending=True )
print(app_sorted)    #answer 8
```

```python
from pandas.tseries.offsets import BMonthEnd
```

```python
apple_month = app['Date'].apply(lambda x: x + BMonthEnd(0))  #mohem
#app.head(5)  #answer 9
apple_month
#The apply method in pandas is a very powerful and flexible tool that allows you to apply a function along any axis of a DataFrame or Series. It takes a function as an argument and applies that function to each element (for a Series) or each row/column (for a DataFrame).
```

```python
apple_month = app.resample('BM').mean()
```

```python
min = app.Date.min()
max = app.Date.max()
print(min, max)
diff = abs(app.Date.max() - app.Date.min()).days  # baraye namayeshe tedade roozha
print(diff)  #answer 10
```
